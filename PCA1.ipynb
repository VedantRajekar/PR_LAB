```python
# PCA on Iris Dataset - Before and After Visualization

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Load dataset
iris = load_iris()
X = iris.data
y = iris.target
feature_names = iris.feature_names
target_names = iris.target_names

# -------- BEFORE PCA (Original Features) --------
plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
for i in range(3):
    plt.scatter(X[y == i, 0], X[y == i, 1], label=target_names[i])
plt.xlabel(feature_names[0])
plt.ylabel(feature_names[1])
plt.title("Before PCA (Original Data)")
plt.legend()

# -------- APPLY PCA --------
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# -------- AFTER PCA --------
plt.subplot(1,2,2)
for i in range(3):
    plt.scatter(X_pca[y == i, 0], X_pca[y == i, 1], label=target_names[i])
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.title("After PCA (Reduced Data)")
plt.legend()

plt.tight_layout()
plt.show()

# Print explained variance
print("Explained Variance Ratio:", pca.explained_variance_ratio_)

```
